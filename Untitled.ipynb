```python
import matplotlib.pyplot as plt 

# x-coordinates of left sides of bars 
left = [1, 2, 3, 4, 5, 6, 7] 

# heights of bars 
height = [1579, 6878, 20, 6771, 20, 20,5932] 

# labels for bars 
tick_label = ['Piracy,Hacking', 'Drugs', 'Fake,Forgery', 'Politics','Hitmen', 'Philosophy','Weapons'] 

# plotting a bar chart 
plt.bar(left, height, tick_label = tick_label, 
		width = 0.1, color = ['red', 'green']) 

# naming the x-axis 
plt.xlabel('Illicits') 
# naming the y-axis 
plt.ylabel('Count') 
# plot title 
plt.title('Count of Illicits from Web Forum 3') 

# function to show the plot 
plt.show() 
```
